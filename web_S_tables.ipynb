```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url="https://ticker.finology.in/"
r=requests.get(url)
print(r) #response 200 means we can scrap the data
```

```python
soup=BeautifulSoup(r.text,"lxml")
print(soup)
```

```python
table=soup.find("table",class_="table table-sm table-hover screenertable")
```

```python
print(table)
```

```python
headers=table.find_all("th")
print(headers)
```

```python
titles=[]
for i in headers:
  title=i.text
  titles.append(title)
print(titles)
```

```python
df=pd.DataFrame(columns=titles)
print(df)
```

```python
rows=table.find_all("tr")
print(rows)
```

```python
for i in rows[1:]:
  print(i.text)
```

```python
for i in rows[1:]:
  data=i.find_all("td")
  row=[tr.text for tr in data]
  #print(row)
  l=len(df)
  df.loc[l]=row
print(df)
```

```python
df.to_csv("stock_market.csv")
```
