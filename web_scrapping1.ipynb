```python
import pandas as pd
```

```python
names=soup.find_all("a",class_ ="title")
print(names)
product_name=[]
```

```python
for i in names:
  name=i.text
  product_name.append(name)
print(product_name)
print(len(product_name))
```

```python
prices=(soup.find_all("h4",{"class":"float-end price card-title pull-right"}))
print(prices)
price_list=[]

```

```python
for i in prices:
  price=i.text
  price_list.append(price)
print(price_list)
print(len(price_list))
```

```python
desc=(soup.find_all("p",class_ ="description card-text"))
print(desc)
desc_list=[]
```

```python
for i in desc:
  des=i.text
  desc_list.append(des)
print(desc_list)
print(len(desc_list))
```

```python
reviews=(soup.find_all("p",class_ ="float-end review-count"))

print(reviews)
rev_list=[]
```

```python
for i in reviews:
  review=i.text
  rev_list.append(review)
print(rev_list)

print(len(rev_list))
```

```python
df=pd.DataFrame({"Product Name":product_name, "Price":price_list,"Description":desc_list, "Number of Reviews":rev_list})
print(df)
```

```python
df.to_csv("pds.csv")
```
